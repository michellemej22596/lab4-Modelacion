### 3. Considere un escenario con 1M de agentes con:

#### a. Versión homogénea: 3 rasgos (idénticos para todos los agentes)

#### b. Versión heterogénea: 5 rasgos únicos por agente

#### c. Responda:

##### i. Calcule y compare los requisitos de memoria para ambas versiones

- **Versión homogénea (3 rasgos):**
  Cada agente tiene 3 rasgos, que son idénticos para todos.
  - Si cada rasgo ocupa 1 unidad de memoria, el total de memoria requerida para los 1M de agentes sería:  
  \( 1,000,000 * 3 \) = **3,000,000 unidades de memoria**.

- **Versión heterogénea (5 rasgos):**
  Cada agente tiene 5 rasgos únicos. Por lo tanto, la memoria requerida será mayor, ya que cada agente tiene información diferente.

  - Si cada rasgo ocupa 1 unidad de memoria, el total de memoria requerida para los 1M de agentes sería:  
  \( 1,000,000 * 5 \) = **5,000,000 unidades de memoria**.

**Comparación:**
- La **versión homogénea** requiere menos memoria (3M unidades de memoria).
- La **versión heterogénea** requiere más memoria (5M unidades de memoria), ya que los rasgos son diferentes para cada agente.

#### ii. Identifique qué rasgos de los agentes podrían simplificarse de forma segura para mejorar el rendimiento sin perder dinámicas críticas

- **Versión heterogénea (5 rasgos):**
  - **Rasgos simplificables:**

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
import numpy as np


def load_data():
    """Returns synthetic discrete-time and continuous-time simulation outputs"""
    np.random.seed(42)  # For reproducibility
    
    # Discrete-time data (weekly reporting)
    discrete = {
        'timestamps': np.arange(0, 21),  # Daily timestamps (integer days)
        'infections': [10, 12, 8, 9, 11, 15, 120,  # Weekly spike
                       9, 11, 7, 8, 10, 125,       # Next week
                       8, 10, 6, 9, 12, 118],      # Final week
        'agent_data': pd.DataFrame({
            'age': np.random.choice(['0-18','19-65','65+'], 1000),
            'occupation': np.random.choice(['healthcare','education','other'], 1000),
            'vaccinated': np.random.choice([True, False], 1000, p=[0.6, 0.4])
        })
    }
    
    # Continuous-time data (event-driven)
    t_continuous = np.linspace(0, 21, 500)
    outbreaks = (80 * np.exp(-(t_continuous-3.5)**2/1.5) + 
                100 * np.exp(-(t_continuous-8.2)**2/2) + 
                90 * np.exp(-(t_continuous-14.1)**2/1.8))
    background = 10 * np.sin(0.3*t_continuous) + 15
    
    continuous = {
        'timestamps': t_continuous,
        'infections': outbreaks + background + np.random.normal(0, 3, 500),
        'agent_data': pd.DataFrame({
            'age': np.random.choice(['0-18','19-65','65+'], 1000),
            'mobility': np.random.gamma(2, 1.5, 1000),  # Continuous trait
            'vaccinated': np.random.choice([True, False], 1000, p=[0.6, 0.4])
        })
    }
    
    return {'discrete': discrete, 'continuous': continuous}
```

```python
# Cargar los datos
data = load_data()

# Datos de simulación de tiempo discreto
discrete_data = data['discrete']
discrete_timestamps = discrete_data['timestamps']
discrete_infections = discrete_data['infections']
discrete_agents = discrete_data['agent_data']

# Datos de simulación de tiempo continuo
continuous_data = data['continuous']
continuous_timestamps = continuous_data['timestamps']
continuous_infections = continuous_data['infections']
continuous_agents = continuous_data['agent_data']
```

```python
# Ajustar el número de elementos para que coincidan
discrete_timestamps = discrete_timestamps[:len(discrete_infections)]

# Ahora puedes graficar
plt.figure(figsize=(10, 5))
plt.plot(discrete_timestamps, discrete_infections, label='Infectados (Tiempo Discreto)', color='blue')
plt.xlabel('Días')
plt.ylabel('Número de Infectados')
plt.title('Simulación de Tiempo Discreto')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(continuous_timestamps, continuous_infections, label='Infectados (Tiempo Continuo)', color='red')
plt.xlabel('Unidades de Tiempo')
plt.ylabel('Número de Infectados')
plt.title('Simulación de Tiempo Continuo')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Inicializar las variables
likely_discrete_time = False
likely_continuous_time = False

# Verificación de si los timestamps son enteros (indicación de tiempo discreto)
if all(isinstance(t, int) for t in discrete_timestamps):
    likely_discrete_time = True
else:
    likely_continuous_time = True

# Imprimir el modelo temporal identificado
print("Modelo temporal: ", "Discreto" if likely_discrete_time else "Continuo")
```

```python
# Buscar picos en el modelo de tiempo discreto
# Picos donde el cambio en las infecciones es máximo (usamos el método de diferencia para detectar picos)
picos_discretos = np.where(np.diff(np.sign(np.diff(discrete_infections))) == -2)[0] + 1

# Verificar si los picos ocurren en intervalos regulares
intervalos_picos = np.diff(discrete_timestamps[picos_discretos])
print("Intervalos entre picos (Tiempo Discreto):", intervalos_picos)

# Si los intervalos son constantes, puede ser evidencia de un modelo de tiempo discreto
if len(set(intervalos_picos)) == 1:
    evidence_for_discrete_time = True
    print("Evidencia de tiempo discreto: Intervalos regulares entre picos.")
```

```python
# Convertir a un array de NumPy para realizar la comparación
infecciones_entre_picos = np.array(discrete_infections[picos_discretos[0]:picos_discretos[-1]])

# Verificar si hay valores bajos entre los picos (usando comparación elemento por elemento)
small_infections_between_peaks = any(infecciones_entre_picos < 5)  # Umbral de 5 infecciones

# Verificar si se cumple la condición
if small_infections_between_peaks:
    evidence_for_continuous_time = True
    print("Evidencia de tiempo continuo: Transmisión de bajo nivel entre picos.")
```

```python
# Calcular la diferencia temporal entre los picos
diferencias_temporales = np.diff(discrete_timestamps[picos_discretos])
print("Diferencias temporales entre picos:", diferencias_temporales)
```

```python
# Inicializar las variables de evidencia
evidence_for_discrete_time = False
evidence_for_continuous_time = False

# Verificar si hay evidencia de tiempo discreto o continuo
if len(set(intervalos_picos)) == 1:
    evidence_for_discrete_time = True
    print("Evidencia de tiempo discreto: Intervalos regulares entre picos.")

if small_infections_between_peaks:
    evidence_for_continuous_time = True
    print("Evidencia de tiempo continuo: Transmisión de bajo nivel entre picos.")

# Evaluar si los patrones observados podrían ser artefactos del modelo temporal
if evidence_for_discrete_time:
    print("Los picos podrían ser artefactos de un modelo temporal discreto.")
elif evidence_for_continuous_time:
    print("La transmisión continua podría estar siendo mal representada.")
```

```python
import numpy as np

# Convertir discrete_infections a un array de NumPy para realizar operaciones matemáticas
discrete_infections = np.array(discrete_infections)

# Agrupar por grupos de edad
age_groups = discrete_agents.groupby('age')['vaccinated'].count()
age_infections = discrete_infections[:len(age_groups)]  

# Calcular las tasas de infección por grupo de edad
age_infection_rate = age_infections / len(discrete_agents)  # Asumimos que la población total es len(discrete_agents)
print("Tasas de infección por grupo de edad:", age_infection_rate)
```

```python
# Agrupar por categorías ocupacionales
occupation_groups = discrete_agents.groupby('occupation')['vaccinated'].count()
occupation_infections = discrete_infections[:len(occupation_groups)]  
# Calcular las tasas de infección por ocupación
occupation_infection_rate = occupation_infections / len(discrete_agents)
print("Tasas de infección por ocupación:", occupation_infection_rate)
```

```python
# Agrupar por estado de vacunación
vaccination_groups = discrete_agents.groupby('vaccinated')['age'].count()
vaccination_infections = discrete_infections[:len(vaccination_groups)]  
# Calcular las tasas de infección por estado de vacunación
vaccination_infection_rate = vaccination_infections / len(discrete_agents)
print("Tasas de infección por estado de vacunación:", vaccination_infection_rate)

```

```python
# Dividir los agentes en vacunados y no vacunados
vaccinated_group = discrete_agents[discrete_agents['vaccinated'] == True]
non_vaccinated_group = discrete_agents[discrete_agents['vaccinated'] == False]

# Calcular la tasa de infección para cada grupo
risk_vaccinated = len(vaccinated_group) / len(discrete_agents)
risk_non_vaccinated = len(non_vaccinated_group) / len(discrete_agents)

# Calcular el riesgo relativo entre los grupos
risk_ratio = risk_vaccinated / risk_non_vaccinated
print(f"Riesgo relativo entre vacunados y no vacunados: {risk_ratio}")
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Convertir los arrays de NumPy a Series de Pandas
age_infection_rate_series = pd.Series(age_infection_rate, index=age_groups.index)
occupation_infection_rate_series = pd.Series(occupation_infection_rate, index=occupation_groups.index)
vaccination_infection_rate_series = pd.Series(vaccination_infection_rate, index=vaccination_groups.index)

# Graficar tasas de infección por grupo de edad
plt.figure(figsize=(10, 5))
age_infection_rate_series.plot(kind='bar', color='skyblue')
plt.title('Tasa de Infección por Grupo de Edad')
plt.xlabel('Grupos de Edad')
plt.ylabel('Tasa de Infección')
plt.show()

# Graficar tasas de infección por ocupación
plt.figure(figsize=(10, 5))
occupation_infection_rate_series.plot(kind='bar', color='salmon')
plt.title('Tasa de Infección por Ocupación')
plt.xlabel('Categorías Ocupacionales')
plt.ylabel('Tasa de Infección')
plt.show()

# Graficar tasas de infección por estado de vacunación
plt.figure(figsize=(10, 5))
vaccination_infection_rate_series.plot(kind='bar', color='lightgreen')
plt.title('Tasa de Infección por Estado de Vacunación')
plt.xlabel('Estado de Vacunación')
plt.ylabel('Tasa de Infección')
plt.show()
```

C. ¿Cómo distinguiría a los verdaderos supercontagiadores de los artefactos de modelado?

Los supercontagiadores son personas que infectan a un número significativamente mayor de individuos que otras, más allá de lo que cabría esperar según las tasas promedio de infección.

Para identificarlos:

- Datos de seguimiento de contacto y transmisión para identificar individuos que hayan infectado a más personas de lo esperado.

- Evaluar si las personas que son supercontagiadores tienen características especiales (por ejemplo, no estar vacunados, vivir en áreas de alta interacción social, etc.).

Los artefactos del modelado pueden ocurrir si el modelo no está correctamente calibrado o si hay suposiciones erróneas (por ejemplo, que todos los agentes tienen la misma probabilidad de infectar a otros). Para identificar artefactos:

- Sensibilidad a cambios en los datos: Si los resultados del modelo cambian drásticamente al mezclar los rasgos o cambiar las ventanas temporales, esto puede indicar artefactos.

- Validación externa: Comparar los resultados con datos del mundo real (por ejemplo, tasas de infección reales) y ajustar el modelo si los supercontagiadores o las tasas de infección no se corresponden con las observaciones.
